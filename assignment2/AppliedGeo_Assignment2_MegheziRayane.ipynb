### 2C) What does this all mean?

A good fit means that the difference between the simulated data and the actual measurement is minimal (commonly calculated by R-squared($data_{measured}-data_{simulated}$ but many other method exist). What is a minimum in this situation will change depending on the needs of the project.

Model 1 has the worst fit and model 3 has the best fit, but it is also the most complex. It is generally better to use simpler models because they limit the number of solutions (as more complex models have more parameters) and since the goal of this inverse problem (in this type of project) is to find the real world parameters of the subsurface, there is only one real solution. For this reason, it is possible and often necessary to use constraints. For example, estimating a possible geometry or other parameter using various surveying techniques would significantly constrain the number of solutions, making the subsurface representation more likely.
